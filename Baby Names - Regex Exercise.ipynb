```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
import os.path

def fetch_babynames(year):
    
    file_name = 'baby'+ str(year)+'.html'
    file_path = os.path.join('E:/aaaML Projects/data/babynames', file_name)

    f = open(file_path, 'r')
    text = f.read()

    # Use brackets around the expressions to easily access/divide into groups for .group() fn.
    year = re.search(r'Popularity\sin\s(\d\d\d\d)', text) 
    final_list = []
    final_list.append(year.group(1))
    names = re.findall(r'<td>(\d+)</td><td>(\w+)</td><td>(\w+)</td>', text)

    rank_list = {}
    for rank, boy, girl in names:
        if boy not in rank_list:
            rank_list[boy] = rank
        if girl not in rank_list:
            rank_list[girl] = rank

    sorted_list = sorted(rank_list.keys())

    for name in sorted_list:
        final_list.append(name + ' ' + rank_list[name])

    print('\n'.join(final_list))
```

```python
fetch_babynames(2006)
```
